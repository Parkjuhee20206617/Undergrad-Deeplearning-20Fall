In [7]:

```
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
    def hello(self):
        print("Hello " + self.name + " !")

    def goodbye(self):
        print("Good-bye " + self.name + " !")
```

In [8]:


```
m = Man("David")
```

In [9]:

```
m.hello()
```

In [10]:

```
m.goodbye()
```
