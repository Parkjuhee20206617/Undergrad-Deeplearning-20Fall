```
import numpy as np
```

```
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    #tmp = np.sum(w*x) + b
    tmp = x[0]*w[0] + x[1]*w[1] + b
    
    if tmp <= 0:
        return 0
    else:
        return 1
```

```
AND(1,1), AND(0,0), AND(1,0), AND(0,1)
```

```
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    #tmp = np.sum(w*x) + b
    tmp = x[0]*w[0] + x[1]*w[1] + b
    
    if tmp <= 0:
        return 0
    else:
        return 1
```

```
NAND(1,1), NAND(0,0), NAND(1,0), NAND(0,1)
```

```
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    #tmp = np.sum(w*x) + b
    tmp = x[0]*w[0] + x[1]*w[1] + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```
OR(1,1), OR(0,0), OR(1,0), OR(0,1)
```

```
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```
XOR(1,1), XOR(0,0), XOR(1,0), XOR(0,1)
```

```
def myOR(x1, x2):
    s1 = NAND(x1, x1)
    s2 = NAND(x2, x2)
    y = NAND(s1, s2)
    return y
```

```
print(myOR(1,1), myOR(0,0), myOR(1,0), myOR(0,1))
print(OR(1,1), OR(0,0), OR(1,0), OR(0,1))
```
